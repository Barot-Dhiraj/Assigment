```python
df = pd.read_csv("drinks.csv")
df #reading CSV file 
```

```python
mean_drink = ['beer_servings', 'spirit_servings','wine_servings','total_litres_of_pure_alcohol'] 
df[mean_drink].mean() #Taking columns into list to find mean
```

```python
std_drinks = mean_drink
df[std_drinks].std()# Standard Deviation
```

```python
var_drinks = std_drinks 
var_drinks_1 = df[var_drinks].var() # finding Variance
var_drinks_1
```

```python
df.describe().round() #Rounding all the math 
```

```python
subset = df.sample(n=20) # here_n_ indicates number varible and the value is 20
subset # Taking all the sample from the population randomly
```

```python
subset.shape # .shape() shows the shape(Rows,Columns) of the subset 
```

```python
rate =  (38) #Sampling at a constant rate
subset = df [::rate] # here :: indicates (Start and Stop) :(start) :(stop) rate(38)
subset # this shows the sample from population with systematic sample (or sample with intervals)
```
